# Here we extracted all the text from given pdf - Page 3

## 1)Synopsis: A brief summary of the objectives and goals of the project described in the PDF.
## 2)Thesis: The main argument or hypothesis that the project aims to test or prove.
## 3)Code: The existing code that you're working with, including any specific sections or functions that are causing errors or need improvement.

```python
import PyPDF2

def extract_text_from_pdf(file_path):
    pdf_file_obj = open(file_path, 'rb')
    pdf_reader = PyPDF2.PdfReader(pdf_file_obj)
    text = ''
    for page_num in range(len(pdf_reader.pages)):
        page_obj = pdf_reader.pages[page_num]
        text += page_obj.extract_text()
    pdf_file_obj.close()
    return text


pdf_text = extract_text_from_pdf(r'C:\Users\trupti kanade\Downloads\Sample_Work\Surender_Gill\Paper-3\3rd.pdf')
print(pdf_text)
```

```python
import spacy

def summarize_objectives_goals(text):
    nlp = spacy.load('en_core_web_sm')
    doc = nlp(text)
    objective_sentences = [sent.text.strip() for sent in doc.sents if "objective" in sent.text.lower() or "goal" in sent.text.lower()]
    summary = ' '.join(objective_sentences)
    return summary

summary = summarize_objectives_goals(pdf_text)
print("Summary of Objectives and Goals:")
print(summary)
```

```python
import spacy

def extract_main_argument_hypothesis(text):
    nlp = spacy.load('en_core_web_sm')
    doc = nlp(text)
    
   
    argument_sentences = [sent.text.strip() for sent in doc.sents if "aim" in sent.text.lower() or "test" in sent.text.lower() or "prove" in sent.text.lower()]
    
    main_argument = argument_sentences[0] if argument_sentences else "Main argument or hypothesis not found."
    
    return main_argument

main_argument = extract_main_argument_hypothesis(pdf_text)
print("Main Argument or Hypothesis:")
print(main_argument)
```

```python
import spacy

def extract_code_info(text):
    nlp = spacy.load('en_core_web_sm')
    doc = nlp(text)
    code_info = []
    for ent in doc.ents:
        if ent.label_ in ['ORG', 'PRODUCT']:  
            if 'code' in ent.text.lower() or 'error' in ent.text.lower() or 'improvement' in ent.text.lower():
                code_info.append(ent.text)
    
   
    main_code_info = ', '.join(code_info) if code_info else "Information about existing code not found."
    
    return main_code_info


code_info = extract_code_info(pdf_text)
print("Information About Existing Code:")
print(code_info)
```
